```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, KFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
```

```python
data_fairuz = pd.read_csv('C:/Users/Fairuz Maulidya/Documents/probstat/houseprices.csv') 
data_fairuz.head(5)
```

```python
col_list=['Price']
numhouse = data_fairuz[data_fairuz.columns[data_fairuz.columns.isin(col_list)]] 
plt.figure(figsize=(10,5))
numhouse.boxplot(sym='r*', grid=False) 
plt.show()
```

```python
plt.figure(figsize=(15,5)) 
plt.subplot(121)
data_fairuz['Price'].plot.hist(bins=10, title='Price') 
plt.show()
```

```python
col_list=['Price', 'SqFt']
numhouse = data_fairuz[data_fairuz.columns[data_fairuz.columns.isin(col_list)]] 
numhouse.plot.scatter(x='SqFt', y='Price')
```

```python
col_list=['Price', 'Bedrooms']
numhouse = data_fairuz[data_fairuz.columns[data_fairuz.columns.isin(col_list)]] 
plt.figure(figsize=(15,5))
numhouse.boxplot(by='Bedrooms') 
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_fairuz['Bedrooms'], data_fairuz['Price'], alpha=0.5)
plt.title('Scatter Plot Harga Rumah Berdasarkan Bedrooms')
plt.xlabel('Jumlah Bedrooms')
plt.ylabel('Harga Rumah')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_fairuz['Bathrooms'], data_fairuz['Price'], alpha=0.5)
plt.title('Scatter Plot Harga Rumah Berdasarkan Bathrooms')
plt.xlabel('Jumlah Bathdrooms')
plt.ylabel('Harga Rumah')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(12, 8))
sns.boxplot(x='Bathrooms', y='Price', data=data_fairuz)
plt.title('Grouped Boxplot Harga Rumah Berdasarkan Bathrooms')
plt.xlabel('Jumlah Bathrooms')
plt.ylabel('Harga Rumah')
plt.xticks(rotation=45)  
# Memutar label x jika diperlukan
plt.grid(True)
plt.show()
```
